```python
# Importing required libraries

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
title_akas_data=pd.read_csv("Dataset/title_akas_data.tsv", sep='\t')
```

**title.akas.tsv** - Contains the following information for titles:

1.   titleId (string) - a tconst, an alphanumeric unique identifier of the title
2.   ordering (integer) – a number to uniquely identify rows for a given titleId
3.   title (string) – the localized title
4.   region (string) - the region for this version of the title
5.   language (string) - the language of the title
6.   types (array) - Enumerated set of attributes for this alternative title. One or more of the following: "alternative", "dvd", "festival", "tv", "video", "working", "original", "imdbDisplay".
7.   attributes (array) - Additional terms to describe this alternative title, not enumerated
8.   isOriginalTitle (boolean) – 0: not original title; 1: original title

Filtering data for India and US Region

```python
title_basics_data=pd.read_csv("Dataset/title_basics_data.tsv", sep='\t')
```

**title.basics.tsv** - Contains the following information for titles:

1.   tconst (string) - alphanumeric unique identifier of the title
2.   titleType (string) – the type/format of the title (e.g. movie, short, tvseries, tvepisode, video, etc)
3.   primaryTitle (string) – the more popular title / the title used by the filmmakers on promotional materials at the point of release
4.   originalTitle (string) - original title, in the original language
5.   isAdult (boolean) - 0: non-adult title; 1: adult title
6.   startYear (YYYY) – represents the release year of a title. In the case of TV Series, it is the series start year
7.   endYear (YYYY) – TV Series end year. ‘\N’ for all other title types
8.   runtimeMinutes – primary runtime of the title, in minutes
9.   genres (string array) – includes up to three genres associated with the title

```python
title_basics_data
```

```python
title_basics_data.isnull().sum()
```

```python
# Renaming titleID column of title_akas_data_India with tconst for merging

title_akas_data_India.rename(columns={'titleId':'tconst'},inplace=True)
```

```python
# Renaming titleID column of title_akas_data_US with tconst for merging

title_akas_data_US.rename(columns={'titleId':'tconst'},inplace=True)
```

```python
merged_title_India=pd.merge(title_akas_data_India,title_basics_data,on=["tconst"])
```

```python
merged_title_US=pd.merge(title_akas_data_US,title_basics_data,on=["tconst"])
```

```python
merged_title_India['titleType'].value_counts()
```

```python
# Filtering data for movies of India region

merged_title_India_movie=merged_title_India[merged_title_India['titleType']=="movie"]
```

```python
# Filtering data for movies of US region

merged_title_US_movie=merged_title_US[merged_title_US['titleType']=="movie"]
```

```python
merged_title_India_movie
```

```python
merged_title_US_movie
```

Replacing "\N" with np.nan ,so that null values can be dropped

```python
merged_title_India_movie.replace('\\N',np.nan,inplace=True)
```

```python
merged_title_US_movie.replace('\\N',np.nan,inplace=True)
```

```python
merged_title_India_movie.isnull().sum()
```

```python
merged_title_US_movie.isnull().sum()
```

```python
merged_title_India_movie=merged_title_India_movie.drop(["types","attributes","endYear"],axis=1)
```

```python
# Have to drop "language" field from US data as 99% of its values are NULL

merged_title_US_movie=merged_title_US_movie.drop(["types","attributes","endYear","language"],axis=1)
```

```python
merged_title_India_movie
```

```python
merged_title_US_movie
```

Dropping Null values from both the dataset

```python
merged_title_India_movie_filter=merged_title_India_movie.dropna()
```

```python
merged_title_US_movie_filter=merged_title_US_movie.dropna()
```

It's time to include rating attributes in our dataset from "title_ratings_data.tsv" file.

```python
title_ratings_data=pd.read_csv("Dataset/title_ratings_data.tsv", sep='\t')
```

**title.ratings.tsv** - Contains the IMDb rating and votes information for titles:

1.   tconst (string) - alphanumeric unique identifier of the title
2.   averageRating – weighted average of all the individual user ratings
3.   numVotes - number of votes the title has received

Merging rating attributes with US and India dataset seperately

```python
US_movie=pd.merge(merged_title_US_movie_filter,title_ratings_data,on=["tconst"])
```

```python
title_crew_data=pd.read_csv("Dataset/title_crew_data.tsv", sep='\t')
```

**title.crew.tsv** - Contains the director and writer information for all the titles in IMDb. Fields include:

1.   tconst (string) - alphanumeric unique identifier of the title
