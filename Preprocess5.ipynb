```python
# Importing reqired libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
actor=pd.read_csv('Dataset/Actors.csv')
```

```python
actress=pd.read_csv('Dataset/Actress.csv')
```

```python
director=pd.read_csv('Dataset/Directors.csv')
```

```python
actor.shape
```

```python
actress.shape
```

```python
director.shape
```

```python
# Merging actor,actess on "tconst"

merged=pd.merge(actor,actress,on=["tconst"])
```

```python
merged
```

```python
# Merging above results with directors_data

merged1=pd.merge(merged,director,on=["tconst"])
```

```python
merged1
```

```python
India_movie=pd.read_csv('Dataset/India_Dataset.csv')
```

```python
India_movie
```

```python
final=pd.merge(India_movie,merged1,on=["tconst"])
```

```python
final
```

```python
# Saving this data to "India_Final.csv".All the analysis for India is to be performed on this dataset.
 
final.to_csv('Dataset/India_Final.csv')
```
